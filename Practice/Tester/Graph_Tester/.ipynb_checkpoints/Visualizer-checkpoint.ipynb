```python
import os
import glob
import sys
```

```python
base_dir = os.getcwd()
# print(base_dir)

kosaraju = open("kosaraju.txt", "r")
tarzan = open("tarzan.txt", "r")
```

```python
vertices = []
edges = []

kosaraju_runtime = []

for line in kosaraju:
    sys.stdout.write(line)
    content = line.split(' ')
    print(content)
    vertices.append(content[0])
    edges.append(content[1])
    kosaraju_runtime.append(content[2])
```

```python
tarzan_runtime = []

for line in tarzan:
    content = line.split(' ')
    tarzan_runtime.append(content[2])
```
